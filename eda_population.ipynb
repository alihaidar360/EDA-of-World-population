```python
sns.histplot(df["Population (2024)"], bins = 10, kde =True)
plt.title("Difference between 2024 Population")
plt.xlabel("data")
plt.ylabel("frequency")
plt.show()
```

```python
Top_5_Countries = df.sort_values(by ="Population (2024)", ascending = False).head(5)
```

```python
sns.barplot(x = "Population (2024)", y="Country", data = Top_5_Countries)
plt.title("Top 5 Countries by Population 2025")
plt.show()
```

```python
df["Population (2024)"] = df["Population (2024)"].replace(",""",regex=False)
sns.scatterplot(data = df, x = "Fert. Rate" , y ="Population (2024)")
```

```python
df.groupby("Country")["Med. Age"].mean().sort_values(ascending = False)
```
